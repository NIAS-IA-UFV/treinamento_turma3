```python
# 2. a)
covidcl.columns = covidcl.columns.str.replace('/','')
chinads = covidcl.loc[covidcl.CountryRegion.isin(['China']),['ProvinceState','Confirmed','Deaths','Recovered']].sort_values(by='Confirmed',ascending = False)
```

```python
# 2. 5 provÃ­ncias com mais casos
# groupby pra juntar provincÃ­as de mesmo nome e .agg('max') pra apenas mostrar o maior valor de casos de cada provÃ­ncia
a = chinads.groupby('ProvinceState').agg('max').sort_values(by='Confirmed',ascending = False).head(5)
display(a)
```

```python
wmd = pd.read_csv("worldometer_data.csv")
cwmd = wmd.loc[:,['Country/Region','Continent','Population','TotalDeaths']]
```

```python
# 4. ranking de mortes por milhÃ£o
deathspermil = cwmd.groupby(['Continent']).sum('TotalDeaths')
deathspermil['Deaths/1MPop'] = ((deathspermil.TotalDeaths/deathspermil.Population) * 10**6)
deathspermil = deathspermil.sort_values(by='Deaths/1MPop',ascending = False)
display(deathspermil)
```
