```python
##Import do data frame
covid = pd.read_csv("../input/corona-virus-report/covid_19_clean_complete.csv")
display(covid)
```

```python
##5.1
##Agrupamento do data frame
covidgb = covid.groupby(['Date','WHO Region'], sort=False).sum('Deaths')
covidgb = covidgb.drop(['South-East Asia', 'Africa', 'Western Pacific'], axis=0,level='WHO Region')
display(covidgb)
```

```python
##Plot do data frame
plot = sns.lineplot(data=covidgb, x="Date", y="Deaths", hue='WHO Region')
##Coloca as grades
plt.grid(True)
plt.title('Numero de mortes Covid-19')
```

```python
##1: Eastern Mediterranean e Europe tiveram mais mortes
##2: Europe antes de 22/03/2020, Americas e Eastern Mediterranen depois de 22/03/2020
##3: Americas: Exponencial de maior base, Eastern Mediterranean: Exponencial de menor base, Europe: Logaritmico
```

```python
##5.2
##Import do data frame do tópico 4.1
wrld = pd.read_csv("../input/corona-virus-report/worldometer_data.csv")
wrld.columns=wrld.columns.str.replace("/","")
wrlddt = wrld[['CountryRegion', 'Continent', 'Population','TotalDeaths']]
k = wrlddt.groupby("Continent").sum("TotalDeaths")
k["Deaths/1MPop"] = ((k.TotalDeaths/k.Population) * 10**6)
k = k.sort_values(by='Deaths/1MPop',ascending = False)
display(k)
```

```python
#Plot do gráfico em barras
bplot = sns.barplot(data = k,x = k.index, y = 'TotalDeaths')
##Titulo
plt.title("Mortes Por Continente")
##Label de x
plt.xlabel('Continente')
##Label de y
plt.ylabel('Mortes')
plt.figure(figsize=(10,6))
```

```python
##1: North America e Europe
##2: Oceania
##3: Pelo tamanho das barras
```

```python
##5.3
##Agrupamento do data frame worldometer_data.csv
fwrld = wrld.groupby('Continent').mean("Deaths/1M pop")
fwrld = fwrld.drop(['Australia/Oceania', 'Asia', 'Africa'])
plt.figure(figsize=(10,6))
##Plot do gráfico de features
sns.scatterplot(data=fwrld, x='Deaths/1M pop', y='Population', hue= 'Continent')
##Título
plt.title("Mortes Por Continente Em Relação À Quantidade De População")
##Label x
plt.xlabel('Mortes') 
##Label y
plt.ylabel('População')
```

```python
##1: South America e Europa
##2: Mesmo com menos população tem muitos casos
##3: Não que eu saiba
```

```python
##5.4
##Plot do mapa de calor
hplot = sns.heatmap(wrld.corr())
##Título
plt.title('Relação')
plt.show(hplot)
##Plot do mapa de calor com features espexíficas
nr = wrld[['Population', 'Deaths/1M pop']]
sns.heatmap(nr.corr())
plt.title('Relação')
```
