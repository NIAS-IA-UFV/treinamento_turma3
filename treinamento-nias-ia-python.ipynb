```python
# 3.1 - Exercício - Twitter Sentiment Analysis

import numpy as np

tweets = [
    "Wow, what a great day today!! #sunshine",
    "I feel sad about the things going on around us. #covid19",
    "I'm really excited to learn Python with @JovianML #zerotopandas",
    "This is a really nice song. #linkinpark",
    "The python programming language is useful for data science",
    "Why do bad things happen to me?",
    "Apple announces the release of the new iPhone 12. Fans are excited.",
    "Spent my day with family!! #happy",
    "Check out my blog post on common string operations in Python. #zerotopandas",
    "Freecodecamp has great coding tutorials. #skillup"
]
```

```python
# Exercício 1
print('Existem',len(tweets),'tweets nesse banco de dados.')
```

```python
# Exercício 2
happy_words = ['great', 'excited', 'happy', 'nice', 'wonderful', 'amazing',
               'good', 'best']
sad_words = ['sad', 'bad', 'tragic', 'unhappy', 'worst']

is_happy = 0
is_sad = 0
pTweet = tweets[0].lower()
for i in happy_words:
    if i in pTweet:
        is_happy += 1

for j in sad_words:
    if j in pTweet:
        is_sad += 1
        
neutro = is_happy-is_sad

if neutro == 0:
    print('O primeiro tweet é neutro.')
elif neutro > 0:
    print('O primeiro tweet é feliz.')
else:
    print('O primeiro tweet não é feliz.')
```

```python
# Exercício 3

nFeliz = 0
nTriste = 0
for t in tweets:
    is_happy = 0
    is_sad = 0
    tweet = t.lower()
    for i in happy_words:
        if i in tweet:
            is_happy += 1

    for j in sad_words:
        if j in tweet:
            is_sad += 1
    
    neutro = is_happy-is_sad

    if neutro > 0:
        nFeliz += 1
    elif neutro < 0:
        nTriste += 1
print('Existem',nFeliz,'tweets felizes e',nTriste,'tweets tristes no BD.')
```
