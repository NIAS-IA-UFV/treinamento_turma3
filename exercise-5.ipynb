```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#5.1#mostrando as regiões que tiveram mais mortes, agrupando por WHO Region e Date
a = pd.read_csv("../input/corona-virus-report/covid_19_clean_complete.csv")
b = a.groupby(['Date','WHO Region'], sort=False).sum('Deaths')
b = b.drop(['South-East Asia', 'Africa', 'Western Pacific'], axis=0,level='WHO Region')
c = sns.lineplot(data=b, x="Date", y="Deaths", hue='WHO Region')
plt.title('Progressão do número de mortes')
plt.grid(True)
c.xaxis.set_major_locator(plt.MaxNLocator(4))
#2#Qual o momento em que cada progressão começa a subir? Eastern Mediterranean e Americas= Mês de Março
#3#Como está a taxa de crescimento de cada progressão? Laranja = exponencial, verde = linear, azul = crescimento muito baixo
plt.show(c)
```

```python
#5.2#Mostra morte por continente agrupando o dataframe por continente e somando as mortes
d = pd.read_csv("../input/corona-virus-report/worldometer_data.csv")
e = d.groupby(['Continent'], sort=False).sum('TotalDeaths')
plt.figure(figsize=(10,6))
f = sns.barplot(data=e,x=e.index, y='TotalDeaths')
plt.title("Morte por continente")
plt.xlabel('Continente') # Eixo x 
plt.ylabel('Mortes totais')
plt.show(f)
#1# Quais os 2 continentes com maior número de mortes? Nort America, Europe
#2# Qual o continente com menor número de mortes? Australia/Oceania
#3# Qual sua hipótese para responder às perguntas anteriores? Ocenia é menor em questão poputacional
```

```python
#5.3#Mostra morte por continente agrupando o dataframe por continente e fazer media das mortes
g = pd.read_csv("../input/corona-virus-report/worldometer_data.csv")
h = g.groupby('Continent').mean("Deaths/1M pop")
h = h.drop(['Australia/Oceania', 'Asia', 'Africa'])
plt.figure(figsize=(10,6))
j = sns.scatterplot(data=h, x='Deaths/1M pop', y='Population', hue= 'Continent')
plt.title("Morte por continente de acordo com a população")
plt.xlabel('Deaths') 
plt.ylabel('População')
plt.show(j)
#1. Onde se concentram a maioria dos dados? Europa, South America
#2. No segundo continentes por mais mortes por milhão, é possível perceber
#alguma relação entre as features Quanto maior a populção mais casos
#3. Pelo gráfico, é possível enxergar relação entre essas colunas do banco de
#dados? não
```

```python
#5.3#Mostra Heatmap de Correlação com base na população e morte
t = pd.read_csv("../input/corona-virus-report/worldometer_data.csv")
v = sns.heatmap(t.corr())
plt.title('Relação')
plt.show(v)
y = t[['Population', 'Deaths/1M pop']]
l = sns.heatmap(y.corr())
plt.title('Relação')
display(l)
```
